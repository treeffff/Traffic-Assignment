```python
import torch
def func(x1, x2):
    """海塞矩阵正定，所以是凸函数"""
    return 2*x1**2 - 4*x1*x2 + 4*x2**2 - 6*x1 - 3*x2

def decay_step_list():
    """生成减速步长"""
    steps = [1/i for i in range(2, 100) for _ in range(i)]
    return steps
```

```python
import torch
x = torch.tensor([-1.0, -1.0])
steps = decay_step_list()
n_step = 0
epslion = 1e-4
while True:
    # 梯度投影
    x = torch.where(x < 0, torch.tensor(0.0), x)
    # 计算梯度
    x.requires_grad_(True)
    f = func(x[0], x[1])
    f.backward()
    grads = x.grad
    x.requires_grad_(False)
    print(f"grads on step{n_step} is {grads}")
    l2_norm = grads[0]**2 + grads[1]**2
    # 如果梯度不再更新，就退出循环
    if (l2_norm - before < epslion):
        remain = remain + 1
    else:
        remain = 0
    if remain > 30:
        print("--------最值出现在边界--------")
        break 
    before = l2_norm
    # 如果梯度为0，就推出循环
    if (l2_norm < epslion):
        break
    # 更新变量
    x = x - grads * steps[n_step]
    n_step += 1
print(f"The result is x:{x}")
```
