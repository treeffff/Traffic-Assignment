```python
import numpy as np
a = np.array([7, 5, 4, 3])
b = np.array([-2, 4, -3, 6])
a_abs = np.sqrt(a@a.T)
b_abs = np.sqrt(b@b.T)
target = a@b.T / (a_abs * b_abs)
target
```

```python
left = 0
right = np.pi
x = (left+right)/2
eplison = 1e-5

def f(x):
    return np.cos(x) - target
 
while abs(f(x)) >= eplison:
    if f(left) * f(x) < 0:
        right = x
    else:
        left = x
    x = (left + right) / 2 
print(f'result: theta={x / np.pi * 180}°')
```

```python
def f2(x):
    """
    函数
    """
    return x**4 - 4*x**3 - 6*x**2 - 16*x + 4

def f_1(x, delte=0.01):
    """
    一阶导
    """
    return (f2(x+delte) - f2(x)) / delte

def f_2(x, delte=0.01):
    """
    二阶导
    """
    return (f_1(x+delte)-f_1(x)) / delte
```

```python
epsilon = 0.001
alpha = 2.0
while abs(f_1(alpha)) >= epsilon:
    alpha -= f_1(alpha) / f_2(alpha)

print(f"result: alpha={alpha}")
```

```python
import numpy as np
import re
def split_string(string):
    # 使用正则表达式匹配文字部分和数字部分
    match = re.match(r'^(.*?)(\(\d+\))$', string)
    
    if match:
        text_part = match.group(1)  # 匹配的文字部分
        number_part = match.group(2)  # 匹配的数字部分
        return text_part, number_part
    else:
        # 如果没有匹配成功，可以返回原始字符串或者抛出异常，视需求而定
        return string, None

# 示例用法
input_string = '电子科学与技术(080900)'
text_part, number_part = split_string(input_string)
print("文字部分:", text_part)
print("数字部分:", type(number_part[1:-1]))
```
